```python
import numpy as np
import matplotlib.pyplot as plt
```

To calculate the values of the coefficients of our linear function, just use the basic equation that determines the model, as well as the cost function. Based on these two equations, we have a direct calculation for the parameters of the function, here called slope and intercept.

---

Para calcular os valores dos coeficientes da nossa função linear, basta usar a equação básica que determina o modelo, assim como a função custo. Com base nessas duas equações, temos um cálculo direto para os parâmetros da função, aqui chamados de inclinação e intercepto.

```python
def calcula_coeficiente(x, y):

    n = np.size(x)
  
    x_media = np.mean(x)
    y_media = np.mean(y)
  
    SS_xy = np.sum(y * x) - (n * y_media * x_media)
    SS_xx = np.sum(x * x) - (n * x_media * x_media)
  
    inclinacao = SS_xy / SS_xx
    intercepto = y_media - inclinacao * x_media
  
    return (intercepto, inclinacao)
```

Now just create the function that generates the graph with the data used and the calculated linear regression.

---

Agora basta criar a função que gera o gráfico com os dados usados e a regressão linear calculada.

```python
def plot_regressao_linear(x, y, b):

    plt.scatter(x, y, color = "m",
               marker = "o", s = 30)
  
    y_pred = b[0] + b[1] * x
  
    plt.plot(x, y_pred, color = "g")
  
    plt.xlabel('x')
    plt.ylabel('y')
  
    plt.show()
```

```python
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)
  
b = calcula_coeficiente(x, y)
print("Coeficientes estimados:\nInclinação = {}\nIntercepto = {}".format(round(b[0], 2), round(b[1], 2)))
  
plot_regressao_linear(x, y, b)
```
